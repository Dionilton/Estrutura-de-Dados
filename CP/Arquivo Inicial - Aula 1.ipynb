# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

#Passo 1: Entrar no sistema da empresa (no caso, link do drive)
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

#demora alguns segundos
time.sleep(7)

#Passo 2: Navegar no sistema e encontrar a base de dados
pyautogui.click(x=374, y=268, clicks = 2)
time.sleep(3)

#Passo 3: Exportar/fazer download da Base de Dados
pyautogui.click(x=380, y=348) #clica no arquivo
pyautogui.click(x=1715, y=159) #clica nos 3 pontinhos
pyautogui.click(x=1506, y=596) #clica em fazer download
time.sleep(5) # Esperar o download


#Passo 4: Importar a base de dados para o Python
#Passo 5: Calcular os indicadores
#Passo 6: Enviar e-mail para diretoria com o relatório
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
#Passo 4: Importar a base de dados para o Python
import pandas as pd

tabela = pd.read_excel(r"C:\Users\dionilton.silva\Downloads\Vendas - Dez.xlsx")
display(tabela)
```

```python
#Passo 5: Calcular os indicadores

faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()

print(faturamento)
print(quantidade)
print(type(tabela))
```

### Vamos agora enviar um e-mail pelo gmail

```python
#Passo 6: Enviar e-mail para diretoria com o relatório

#abrir meail(https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox)
pyautogui.hotkey("ctrl","t")
pyperclip.copy("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(10)

#clica no escrever
pyautogui.click(x=145, y=158)
time.sleep(8)

#escreve email destinatarui
pyautogui.write("dionilton.os@gmail.com")
pyautogui.press("tab") #seleciona email 1
pyautogui.write("henniton@autlook.com") 
pyautogui.press("tab") #seleciona email 2
pyautogui.press("tab") #passa para o campo assunto

#escreve assunto
pyperclip.copy("Teste de automação com Python")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") #passa para o corpo do email

#escreve corpo email
texto = f"""
Bom dia!

E-mail teste de automação com python

O faturamento de ontem foi: {faturamento:,.2f}
A quantidade de produtos foi: {quantidade:,}

Pode excluir depois da sua caixa de entrada

atenciosamente,
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

#enviar email
pyautogui.hotkey("ctrl", "enter")
```
